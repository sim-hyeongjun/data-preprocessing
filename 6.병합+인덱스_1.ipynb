#### 함수 문제 리뷰 주민번호

```python
def check_jumin():
    jumin_str = input("주민번호 입력(000000-0000000):")
    jumin_nums = []
    for s in jumin_str:
        if s =="-" :
            continue
        else :
            jumin_nums.append(int(s))
    digit=2
    total = 0
    for idx in range(len(jumin_nums)-1) :
        if digit == 10 :
            digit=2
        total += jumin_nums[idx]*digit
        digit+=1
    check_digit = (11-total%11)%10
    if check_digit == jumin_nums[-1] :
        print("CORRECT")
    else :
        print("INCORRECT")
```

```python
import random
count  = 0
serial = 0
while True :
    dices = set()
    count +=1
    for i in range(6):
        dices.add(random.randint(1, 7))
    if len(dices) == 6 :
        serial +=1
    elif len(dices) == 1:
        break
```
