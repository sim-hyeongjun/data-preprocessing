```python
#pandas의 Series클래스는 numpy의 1차원 배열 객체와 유사한 데이터 구조 객체 생성
#pandas의 Series 객체는 데이터값 + index(정수 또는 label문자열)
#Series 객체[정수인덱스]
#Series 객체[label인덱스]
#Series 객체.index
#Series 객체.values
s0 = pd.Series([9904312,3448737,2890451,2466052])
print('s0.index: ',s0.index)
print()
print('s0.valuse: ', s0.values)
s = pd.Series([9904312,3448737,2890451,2466052], index=['서울', '부산', '인천', '대구'])
print('s.index: ', s.index)
print()
print('s.valuse: ', s.values)
s.name = '인구' #데이터 이름
s.index.name = '도시' #lable인덱스의 이름
print()
print('s: ',s)
print()
print(s/100000) #broadcast되어서 vectorized operation 지원
# 정수인덱스, label인덱스 모두 indexing, slicing 지원됨
print()
print('s[:3] ,s[부산] : ',s[3], s['부산'])
print()
print('s[:3] : ',s[:3], s['부산':'대구'])
```

```python
print(type(s[:3])) #슬라이싱으로 추출한 부분집합의 반환 타입?
print(s.부산) # Series객체.label인덱스로 데이터 요소 처리 가능
print('부산' in s) # True
print('광주' in s) # False
```

```python
#dict로부터 (key, value) 로 추출하려면..items()
for v,k in s.items() :
    print('%s = %d'%(v,k))
```
