```python
#pclass 기준으로 그룹화

tanic_group=tanic__df.groupby(by='Pclass')
#객실별 인원
print(tanic_group['Pclass'].count())
#그룹바이데이터 전체 개수
print(tanic_group.count())

#객실등급별 나이와 fare 평균
print(tanic_group[['Age', 'Fare']].mean())

#객실별 최고령 , 최연소
print(tanic_group['Age'].agg((max, min)))

#평균치
print(tanic__df[['Age', 'Fare']].mean())
```

```python
# 객실별 최고령 , subsp 최저 , fare 평균
#서로 다른 컬럼에 서로 다른 aggregation 메소드를 적용할 경우 agg()내에 컬럼과 적용할 메소드를 Dict형태로 입력
agg_format= {'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
print(tanic__df.groupby('Pclass').agg(agg_format))
# 같은 방법
tanic__df.groupby(['Pclass']).agg(age_max=('Age', 'max'), age_mean=('Age', 'mean'), fare_mean=('Fare', 'mean'))

tanic__df.groupby(['Pclass']).agg(agg_max=('Age', 'min'), age_mean=('Age', 'min'), fare_mean=('SibSp','sum'))
```
